```python
image = Image.open(my_path)
```

```python
# Chuyen doi hinh anh thanh mang NumPy
image_array = np.array(image.convert("L"))

# Bien doi anh theo yeu cau
inverted_image = 300 - image_array  # Invert anh bang cach lay 300 gtri pixel 
squared_image = image_array ** 2  # Binh phuong gia tri pixel
clamped_image = np.clip(image_array, 100, 200)  # Gioi han gia tri pixel

#Hien thi hinh anh goc va hinh anh da bien doi voi khoang trang
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.5, hspace=0.5)

axs[0, 0].imshow(image_array, cmap="gray")
axs[0, 0].set_title('Anh Goc')

axs[0, 1].imshow(inverted_image, cmap="gray")
axs[0, 1].set_title('Anh da bien doi')

axs[1, 0].imshow(squared_image, cmap="gray")
axs[1, 0].set_title('Anh da bien doi')

axs[1, 1].imshow(clamped_image, cmap="gray")
axs[1, 1].set_title('Anh da bien doi')

plt.show()
```
