```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#duong dan anh
my_path= 'C:/Users/ducth/Documents/hoctap/thigiacmaytinh/img/ava_spiderman.png'
#doc anh
img = Image.open(my_path)
plt.imshow(img)
```

```python
#chuyen sang anh xam
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array =np.array(img_gray)
print(img_array)
```

```python
#tao mot hinh ve moi
plt.figure()
plt.hist(img_array.flatten(), bins=32)
plt.title('Bieu do histogram cua anh')
plt.xlabel('Gia tri cua pixel')
plt.ylabel('So luong')
plt.show()
```

```python
#highlight duong bien
plt.figure()
plt.gray()
plt.contour(img_gray, origin='image')
plt.show()
```
