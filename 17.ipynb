```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image_path = 'C:/Users/ducth/Documents/hoctap/thigiacmaytinh/img/ava_spiderman.png'
image = Image.open(image_path)
```

```python
gray_image = image.convert('L')
```

```python
image_a = np.array(gray_image)
```

```python
# Tinh gradient theo huong x
gradient_x = ndimage.sobel(image_a, axis=0, mode='constant')
```

```python
# Hien thi anh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Anh xam goc')

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Anh')

plt.show()
```

```python
gradient_x = ndimage.sobel(image_a, axis=1, mode='constant')
```

```python
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Anh xam goc')

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Anh xam da bien doi')

plt.show()
```

```python
# Tinh gradient theo huong x (dao ham rieng theo x) bang mat na Prewitt x
gradient_x = ndimage.convolve(image_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
# Tinh gradient theo huong y (dao ham rieng theo y) bang mat na Prewitt y
gradient_y = ndimage.convolve(image_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
# Tinh bien do gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Hien thi anh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Anh xam goc')

plt.subplot(2,1,2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Anh da bien doi')

plt.show()
```
