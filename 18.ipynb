```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image
```

```python
image = Image.open('C:/Users/ducth/Documents/hoctap/thigiacmaytinh/img/car.jpg')
```

```python
gray_image = image.convert('L')
image_array = np.array(gray_image)
```

```python
#chuyen sang anh nhi phan
binary_image = 1*(image_array<200)
```

```python
# Danh nhan
labeled_image, num_features = ndimage.label(binary_image)

# Tao luoi subplot voi 3 dong 1 cot
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Anh xam")

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Anh nhi phan")

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Da danh nhan ({num_features} doi tuong)")

plt.show();
```

```python
# Thuc hien opening de loai bo nhieu va noi doi tuong
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((5,5)))

# Danh nhan
labeled_image, num_features = ndimage.label(opened_image)

# Tao luoi subplot
fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Anh xam")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Anh nhi phan")
                                      
plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("Anh sau opening")


plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Da danh nhan~ ({num_features} Doi tuong)")

plt.show();
```
